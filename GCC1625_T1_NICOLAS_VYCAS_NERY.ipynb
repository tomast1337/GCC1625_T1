# 1 Teorema do Limite Central

Nesta seção, exploraremos o Teorema do Limite Central (TLC) utilizando a distribuição exponencial. O objetivo é realizar simulações computacionais e verificar empiricamente como a distribuição das médias amostrais se aproxima de uma distribuição normal, conforme enunciado no TLC.

## Objetivos
1. Gerar uma amostra de tamanho 1000 da distribuição exponencial com parâmetro λ = 0.2 e criar um histograma.
2. Criar um histograma da distribuição amostral das médias para 1000 simulações de tamanho 40.
3. Calcular a média e a variância da distribuição amostral das médias e compará-las com os valores teóricos.

## Teorema do Limite Central
O Teorema do Limite Central afirma que, para amostras suficientemente grandes, a distribuição amostral da média de variáveis independentes e identicamente distribuídas se aproxima de uma distribuição normal, independentemente da forma da distribuição original.

### Função de Densidade da Distribuição Exponencial
A densidade de probabilidade para a variável aleatória \(X\), que segue uma distribuição exponencial, é definida como:
\[
f(x; \lambda) =
\begin{cases}
\lambda e^{-\lambda x}, & \text{se } x \geq 0, \\
0, & \text{se } x < 0.
\end{cases}
\]

- **Média da distribuição exponencial:** \( \mu = \frac{1}{\lambda} \)  
- **Desvio padrão:** \( \sigma = \frac{1}{\lambda} \)  

No nosso caso, \( \lambda = 0.2 \).

### Procedimentos
1. **Geração de uma amostra exponencial:** Utilizamos a função `expon.rvs` da biblioteca `scipy.stats` para gerar os dados.
2. **Histograma da amostra original:** Visualização da distribuição exponencial.
3. **Simulações para distribuição amostral das médias:** Realização de 1000 simulações de amostras de tamanho 40.
4. **Cálculo e comparação dos valores:** Verificação de como a média e variância calculadas se aproximam dos valores teóricos.

### Resultados
Os gráficos e análises serão apresentados nas células a seguir.


### Código 1: Geração de uma amostra exponencial

```python
# Importando bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
import rich as r
print = r.print

# Parâmetros
lmbda = 0.2  # Parâmetro lambda da distribuição exponencial
sample_size = 1000  # Tamanho da amostra

# Gerando a amostra exponencial
sample = expon.rvs(scale=1/lmbda, size=sample_size)

# Plotando o histograma da amostra
plt.hist(sample, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma da Amostra Exponencial')
plt.xlabel('Valores da Amostra')
plt.ylabel('Frequência')
plt.show()
```

#### Geração de uma Amostra Exponencial

No código acima:
1. Definimos o parâmetro da distribuição exponencial (\( \lambda = 0.2 \)).
2. Geramos uma amostra de tamanho 1000 utilizando a função `expon.rvs` da biblioteca `scipy.stats`.
3. Criamos um histograma para visualizar a distribuição exponencial gerada.

O histograma mostra a assimetria característica da distribuição exponencial.


### Código 2: Geração da distribuição amostral das médias

```python
# Parâmetros para simulação
n = 40  # Tamanho de cada amostra
simulations = 1000  # Número de simulações

# Calculando a média amostral para cada simulação
sample_means = [np.mean(expon.rvs(scale=1 / lmbda, size=n)) for _ in range(simulations)]

# Plotando o histograma da distribuição amostral das médias
plt.hist(sample_means, bins=30, color="lightgreen", edgecolor="black", density=True)
plt.title("Histograma da Distribuição Amostral das Médias")
plt.xlabel("Média Amostral")
plt.ylabel("Densidade")
plt.show()
```

#### Geração da Distribuição Amostral das Médias

No código acima:
1. Definimos um tamanho de amostra (\(n = 40\)) e o número de simulações (\(1000\)).
2. Calculamos a média de cada uma das 1000 amostras, utilizando uma lista de compreensões em Python.
3. Plotamos um histograma para a distribuição das médias amostrais.

De acordo com o Teorema do Limite Central, a distribuição das médias amostrais se aproxima de uma normal, independentemente da distribuição original.


### Código 3: Cálculo e comparação de estatísticas teóricas e empíricas

```python
# Estatísticas teóricas
mean_theoretical = 1 / lmbda  # Média teórica da distribuição exponencial
variance_theoretical = (1 / lmbda) ** 2 / n  # Variância teórica das médias amostrais

# Estatísticas empíricas
mean_sample = np.mean(sample_means)  # Média das médias amostrais
variance_sample = np.var(sample_means)  # Variância das médias amostrais

# Exibindo os resultados
print(f"Média Teórica: {mean_theoretical:.4f}, Média Amostral: {mean_sample:.4f}")
print(
    f"Variância Teórica: {variance_theoretical:.4f}, Variância Amostral: {variance_sample:.4f}"
)
```

#### Cálculo e Comparação de Estatísticas

No código acima:

2. Obtivemos os valores empíricos calculados a partir das 1000 simulações: a média e a variância das médias amostrais.
3. Comparamos os valores teóricos com os empíricos para verificar a aderência aos resultados esperados pelo Teorema do Limite Central.

#### (i): Histograma da Distribuição da População

```python
import numpy as np
import matplotlib.pyplot as plt

# População modelada como uma distribuição uniforme discreta
population = np.arange(5, 16)  # Valores de 5 a 15

# Histograma da distribuição populacional
plt.bar(
    population,
    [1 / len(population)] * len(population),
    color="skyblue",
    edgecolor="black",
)
plt.title("Distribuição Populacional (Quantidade de Batatas)")
plt.xlabel("Quantidade de Batatas")
plt.ylabel("Probabilidade")
plt.xticks(population)
plt.show()
```

#### Explicação:

- A população é composta pelos inteiros de 5 a 15.
- A probabilidade de cada valor é 1/11, já que todos têm a mesma chance (distribuição uniforme discreta).

#### (ii): Cálculo de Média, Variância e Desvio Padrão

```python
# Estatísticas da população
mean_population = np.mean(population)  # Média
variance_population = np.var(population, ddof=0)  # Variância (populacional)
std_population = np.sqrt(variance_population)  # Desvio padrão

print(f"Média da População (μ): {mean_population}")
print(f"Variância da População (σ²): {variance_population}")
print(f"Desvio Padrão da População (σ): {std_population}")
```

#### Explicação

- A **média** é calculada somando todos os valores da população e dividindo pelo total:
  \[
  \mu = \frac{\sum x_i}{N}
  \]

- A **variância populacional** (\( \sigma^2 \)) é dada por:
  \[
  \sigma^2 = \frac{1}{N} \sum (x_i - \mu)^2
  \]
  Onde:
  - \( x_i \) são os valores individuais da população.
  - \( \mu \) é a média populacional.
  - \( N \) é o tamanho da população.

- O **desvio padrão** (\( \sigma \)) é a raiz quadrada da variância:
  \[
  \sigma = \sqrt{\sigma^2}
  \]


#### (iii): Distribuição Amostral da Média

- A média de cada amostra é calculada.
- O histograma mostra como as médias das amostras estão distribuídas.


#### (iv): Estatísticas da Distribuição Amostral

```python
# Estatísticas da distribuição amostral da média
mean_sample_dist = np.mean(sample_means)
variance_sample_dist = np.var(sample_means, ddof=0)  # Variância
std_sample_dist = np.sqrt(variance_sample_dist)  # Desvio padrão

print(f"Média da Distribuição Amostral: {mean_sample_dist}")
print(f"Variância da Distribuição Amostral: {variance_sample_dist}")
print(f"Desvio Padrão da Distribuição Amostral: {std_sample_dist}")
```

  - \( \sigma^2 \) é a variância populacional.

- O **desvio padrão teórico** da distribuição amostral é:
  \[

# 3 Distribuição Amostral da Diferença de Médias

Nesta seção, estudaremos a diferença entre as médias de duas populações hipotéticas:
1. **População 1:** Pessoas que tomaram o medicamento.
   - Média (\( \mu_1 \)) = 50
   - Variância (\( \sigma_1^2 \)) = 25
2. **População 2:** Pessoas que não tomaram o medicamento.
   - Média (\( \mu_2 \)) = 40
   - Variância (\( \sigma_2^2 \)) = 24

A diferença entre as médias populacionais é \( \mu_{Md} = \mu_1 - \mu_2 = 10 \).

### Fórmulas Teóricas
1. **Variância da diferença de médias**:
   \[
   \sigma^2_{Md} = \frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}
   \]
2. **Erro padrão da diferença de médias**:
   \[
   \sigma_{Md} = \sqrt{\sigma^2_{Md}}
   \]

No exemplo:
- \( n_1 = 10 \), \( n_2 = 8 \)
- \( \sigma^2_{Md} = \frac{25}{10} + \frac{24}{8} = 5.5 \)
- \( \sigma_{Md} = \sqrt{5.5} \approx 2.35 \)


#### (i): Histograma da Diferença de Médias

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
mu1, sigma1_squared = 50, 25  # População 1
mu2, sigma2_squared = 40, 24  # População 2
n1, n2 = 10, 8  # Tamanhos das amostras
simulations = 10000  # Número de simulações

# Gerando a distribuição amostral da diferença de médias
differences = []
for _ in range(simulations):
    sample1 = np.random.normal(mu1, np.sqrt(sigma1_squared), n1)
    sample2 = np.random.normal(mu2, np.sqrt(sigma2_squared), n2)
    M1 = np.mean(sample1)
    M2 = np.mean(sample2)
    differences.append(M1 - M2)

# Histograma
plt.hist(differences, bins=30, color="skyblue", edgecolor="black", density=True)
plt.title("Distribuição Amostral da Diferença de Médias (Md)")
plt.xlabel("Diferença de Médias (Md)")
plt.ylabel("Densidade")
plt.show()
```

#### Explicação

- Geramos 10.000 simulações de amostras para as populações 1 e 2.
- Para cada simulação, calculamos as médias \( M_1 \) e \( M_2 \), e a diferença \( M_d = M_1 - M_2 \).
- O histograma mostra a distribuição amostral da diferença de médias.


#### (ii): Cálculo da Média e do Desvio Padrão

```python
# Estatísticas da distribuição amostral
mean_difference = np.mean(differences)
std_difference = np.std(differences)

print(f"Média Empírica (Md): {mean_difference}")
print(f"Desvio Padrão Empírico (Md): {std_difference}")

# Valores teóricos
theoretical_mean = mu1 - mu2
theoretical_variance = sigma1_squared / n1 + sigma2_squared / n2
theoretical_std = np.sqrt(theoretical_variance)

print(f"Média Teórica (Md): {theoretical_mean}")
print(f"Desvio Padrão Teórico (Md): {theoretical_std}")
```

#### Explicação

- A média empírica é calculada diretamente da lista `differences`.
- O desvio padrão empírico é comparado com o valor teórico calculado pelas fórmulas fornecidas.
- A proximidade entre valores empíricos e teóricos valida o modelo.


#### (iii): Probabilidade de Md ≥ 15

```python
from scipy.stats import norm

# Parâmetros da distribuição normal aproximada
mu_Md = theoretical_mean
sigma_Md = theoretical_std

# Probabilidade de Md >= 15
p_md_greater_15 = 1 - norm.cdf(15, loc=mu_Md, scale=sigma_Md)

print(f"Probabilidade de Md >= 15: {p_md_greater_15:.4f}")
```

#### Explicação

- Utilizamos a função de distribuição acumulada (CDF) da distribuição normal para calcular a probabilidade.
- A probabilidade \( P(M_d \geq 15) \) é a área sob a curva normal à direita de 15.

## 4  Intervalo de confiança para média populacional

A pesquisa de Schroeder e Epley (2015) avaliou se recrutadores percebem candidatos de forma diferente dependendo de como suas habilidades são apresentadas: via áudio (fala) ou transcrição (texto). O objetivo é comparar as avaliações de intelecto em duas condições (Sa e St), analisar os dados estatisticamente e verificar a validade das conclusões dos autores.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, t
import numpy as np

# Carregar os dados
data = pd.read_csv("SchroederEpley2015data.csv", sep=",")

data.head()
```

```python
# Separar as amostras
sa = data[data["CONDITION"] == 1]["Intellect_Rating"]
st = data[data["CONDITION"] == 0]["Intellect_Rating"]
```

```python
# (i) Média e tamanho das amostras
mean_sa = sa.mean()
mean_st = st.mean()
size_sa = sa.size
size_st = st.size

print(f"Média Sa: {mean_sa}, Tamanho Sa: {size_sa}")
print(f"Média St: {mean_st}, Tamanho St: {size_st}")
```

#### (i) Média e tamanho das amostras
Para cada condição (Sa e St), calculamos a média e o número de observações. Essas informações ajudam a entender as diferenças iniciais entre os grupos.

```python
# (ii) Boxplot
sns.boxplot(x=data["CONDITION"], y=data["Intellect_Rating"], palette="Set2", hue=data["CONDITION"])
plt.xlabel("CONDITION")
plt.ylabel("Intellect Rating")
plt.title("Boxplot para Sa e St")
plt.xticks([0, 1], ["Transcrição", "Áudio"])
plt.show()
```

#### (ii) Boxplot
O boxplot exibe graficamente a distribuição das avaliações em ambas as condições, permitindo identificar tendências centrais e dispersões.

```python
# (iii) Teste de normalidade (Shapiro-Wilk)
shapiro_sa = shapiro(sa)
shapiro_st = shapiro(st)

print(
    f"Teste de Shapiro para Sa: estatística={shapiro_sa.statistic}, p-valor={shapiro_sa.pvalue}"
)
print(
    f"Teste de Shapiro para St: estatística={shapiro_st.statistic}, p-valor={shapiro_st.pvalue}"
)
```

#### (iii) Teste de normalidade
Realizamos o teste de Shapiro-Wilk para verificar se as distribuições das amostras são normais.

```python
# (iv) Intervalos de confiança (nível de 99%)
confidence_level = 0.99
alpha = 1 - confidence_level


# Função para calcular o intervalo de confiança
def confidence_interval(data, confidence):
    n = len(data)
    mean = data.mean()
    std_err = data.std(ddof=1) / np.sqrt(n)
    t_value = t.ppf(1 - alpha / 2, n - 1)
    margin_of_error = t_value * std_err
    return mean - margin_of_error, mean + margin_of_error


ci_sa = confidence_interval(sa, confidence_level)
ci_st = confidence_interval(st, confidence_level)

print(f"Intervalo de confiança para Sa (áudio): {ci_sa}")
print(f"Intervalo de confiança para St (transcrição): {ci_st}")
```

#### (iv) Intervalos de confiança

Para cada condição, construímos intervalos de confiança para a média com um nível de 99%. Esses intervalos fornecem uma estimativa do verdadeiro valor médio da população.

```python
# (v) Análise final
if ci_sa[0] > ci_st[1] or ci_st[0] > ci_sa[1]:
    conclusion = "Os intervalos de confiança não se sobrepõem. A conclusão dos autores parece válida."
else:
    conclusion = "Os intervalos de confiança se sobrepõem. A conclusão dos autores pode não ser válida."

print(conclusion)
```

#### (v) Análise final

Os intervalos de confiança não se sobrepõem, indicando que os recrutadores avaliam candidatos de forma significativamente melhor na condição de áudio (Sa) em comparação com a transcrição (St).

## 5 Intervalo de Confiança para Média Populacional

Nesta seção, vamos analisar um estudo realizado por Schroeder e Epley (2015) sobre como o formato de um discurso (áudio ou transcrição) influencia as percepções de um recrutador sobre as habilidades de um candidato. O objetivo é construir intervalos de confiança para as médias de avaliações de recrutadores nas condições de áudio e transcrição.

O conjunto de dados utilizado contém duas colunas de interesse:
- **CONDITION**: Indica a condição à qual cada recrutador foi alocado. O valor 1 indica que o recrutador foi alocado na condição de áudio; 0 indica que o recrutador foi alocado na condição de transcrição.
- **Intellect_Rating**: A avaliação atribuída ao candidato, com valores entre 0 e 10.

Com base nessas informações, responderemos aos seguintes itens.


```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Carregar os dados
df_survey = pd.read_csv("survey.csv")
df_survey.head()
```

#### (i) Usando a distribuição t de Student, calcule um intervalo de confiança no nível de 95% para a altura média dos estudantes da universidade.

```python
# Remover valores ausentes na coluna 'Height'
df_survey = df_survey.dropna(subset=["Height"])

# Calcular a média e o desvio padrão da amostra
mean_height = df_survey["Height"].mean()
std_dev = df_survey["Height"].std()
n = len(df_survey["Height"])

# Erro padrão da média
se = std_dev / np.sqrt(n)

# Valor crítico t para 95% de confiança (graus de liberdade = n - 1)
t_critical = stats.t.ppf(1 - 0.025, df=n - 1)

# Intervalo de confiança
margin_of_error = t_critical * se
ci_lower = mean_height - margin_of_error
ci_upper = mean_height + margin_of_error

# Exibir o intervalo de confiança
print(f"Intervalo de Confiança (t de Student): ({ci_lower:.2f}, {ci_upper:.2f})")
```

#### Intervalo de Confiança com t de Student (nível de confiança de 95%)

Para calcular o intervalo de confiança usando a distribuição **t de Student**, seguimos os seguintes passos:

1. **Cálculo da média e do desvio padrão da amostra**:
   - A média (\( \bar{x} \)) e o desvio padrão (\( s \)) são calculados a partir dos dados.

2. **Cálculo do erro padrão da média**:
   - O erro padrão (\( SE \)) é dado por:
     \[
     SE = \frac{s}{\sqrt{n}}
     \]
     onde \( s \) é o desvio padrão da amostra e \( n \) é o tamanho da amostra.

3. **Valor crítico t**:
   - Utilizamos a função `stats.t.ppf` para determinar o valor crítico \( t_{\alpha/2} \), considerando:
     - Nível de confiança de 95%.
     - Graus de liberdade \( n - 1 \).

4. **Cálculo do intervalo de confiança**:
   - O intervalo de confiança é dado por:
     \[
     IC = \bar{x} \pm t_{\alpha/2} \times SE
     \]


#### (ii) Construa outro intervalo de confiança, desta vez usando o z-score (em vez do t-score que você usou anteriormente).

```python
# Erro padrão da média (já calculado no passo anterior)
# Valor crítico z para 95% de confiança (z = 1.96)
z_critical = 1.96

# Intervalo de confiança com z-score
margin_of_error_z = z_critical * se
ci_lower_z = mean_height - margin_of_error_z
ci_upper_z = mean_height + margin_of_error_z

# Exibir o intervalo de confiança com z-score
print(f"Intervalo de Confiança (z-score): ({ci_lower_z:.2f}, {ci_upper_z:.2f})")
```

#### Cálculo do intervalo com z-Score:

Para o cálculo do intervalo de confiança com o z-score, usamos o valor crítico 𝑧 =1.96, que corresponde a um nível de confiança de 95% para distribuições normais.
O cálculo é feito da mesma forma que para o t de Student, mas usando o valor fixo de 𝑧.

#### (iii) Apresente uma análise comparativa dos dois intervalos de confiança obtidos.

```python
# Comparar as larguras dos intervalos de confiança
width_t = ci_upper - ci_lower
width_z = ci_upper_z - ci_lower_z

print(f"Largura do Intervalo de Confiança (t de Student): {width_t:.2f}")
print(f"Largura do Intervalo de Confiança (z-score): {width_z:.2f}")

# Discussão: Analisar qual intervalo é mais estreito e por quê.
if width_t < width_z:
    print(
        "O intervalo de confiança com t de Student é mais estreito, indicando que a variância da amostra foi bem estimada."
    )
else:
    print(
        "O intervalo de confiança com z-score é mais estreito, o que pode ocorrer devido à suposição de variância conhecida."
    )
```

#### Análise Comparativa:

Comparamos as larguras dos dois intervalos para entender a precisão de cada método.
O intervalo de confiança com t de Student tende a ser mais amplo em amostras pequenas ou quando a variância da população não é conhecida.
